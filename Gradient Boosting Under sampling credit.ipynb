```python
dfinvalid=dfcredit[dfcredit["Class"]==1]
```

```python
dfvalid=dfcredit[dfcredit["Class"]==0]
```

```python
dfval=dfvalid.sample(492)
```

```python
dfval
```

```python
dffinal=pd.concat([dfval,dfinvalid])
```

```python
dffinal
```

```python
x=dffinal.drop("Class",axis=1)
```

```python
y=dffinal["Class"]
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.20)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc=GradientBoostingClassifier()
```

```python
gbc.fit(xtrain,ytrain)
```

```python
gbc.score(xtest,ytest)
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
kf=KFold(n_splits=10)
```

```python
gda=GradientBoostingClassifier()
```

```python
score=cross_val_score(gda,x,y,cv=kf)
```
