```python
from sklearn.model_selection import train_test_split
```

```python
x2train,x2test,y2train,y2test=train_test_split(x2,y2,test_size=.20)
```

```python
x2train.shape,x2test.shape,y2train.shape,y2test.shape
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
grd=GradientBoostingClassifier()
```

```python
grd.fit(x2train,y2train)
```

```python
grd.score(x2test,y2test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada=AdaBoostClassifier()
```

```python
ada.fit(x2train,y2train)
```

```python
ada.score(x2test,y2test)
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
kfold=KFold(n_splits=5)
```

```python
gda1=GradientBoostingClassifier()
```
